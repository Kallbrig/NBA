```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
james_per = pd.read_csv('Stats/James/Playoffs/Per Game - Playoffs.csv')
jordan_per = pd.read_csv('Stats/Jordan/Playoffs/Per Game - Playoffs.csv')
james_adv = pd.read_csv('Stats/James/Playoffs/Per Game Advanced - Playoffs.csv')
jordan_adv = pd.read_csv('Stats/Jordan/Playoffs/Per Game Advanced - Playoffs.csv')

james = pd.merge(james_per,james_adv,on=['Season','Age'])
jordan = pd.merge(jordan_per,jordan_adv,on=['Season','Age'])
```

```python
james = james.iloc[0:5]
james
```

```python
jordan = jordan.iloc[:6]
jordan
```

```python
for i in enumerate(['PTS','MP_x','AST','TRB','STL','BLK','TOV','TS%','OWS','DWS','WS/48','OBPM','DBPM','BPM','VORP','PER','USG%']):
    plt.style.use('fivethirtyeight')

    plt.title(i[1])
    plt.plot(james[i[1]],label='James')
    plt.plot(jordan[i[1]],label='Jordan')
    plt.legend()
    plt.tight_layout()

    if i[1] == 'WS/48':
        plt.savefig(f'{i[0]} Playoffs - Individual Performance WS_48',dpi=500)
    else:
        plt.savefig(f'{i[0]} Playoffs - Individual Performance {i[1]}',dpi=500)
    plt.show()
```

```python
for i in ['PTS','MP_x','AST','TRB','STL','BLK','TOV','TS%','OWS','DWS','WS/48','OBPM','DBPM','BPM','VORP','PER','USG%']:
    print(i,)
    print('James -',round(james.loc[:,i].mean(),2))
    print('Jordan -',round(jordan.loc[:,i].mean(),2))
    print()
```
