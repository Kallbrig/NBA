```python
import pandas as pd
import requests

from selenium import webdriver
import time
from selenium.webdriver.chrome.options import Options
```

```python
df = pd.read_csv('../Basketball Reference Stat Scraper/all/all.csv',index_col=0)
```

```python
# link for extract html data
def getdata(url):
    r = requests.get(url)
    return r.text

```

```python
df['Name'] = df.Player.str.strip('*')
```

```python
lebron = df[df['Name'] == 'LeBron James']
jordan = df[df['Name'] == 'Michael Jordan']
```

```python
table_list = ['Per Game',
 'Per Game Playoffs',
 'Totals',
 'Totals Playoffs',
 'Per 36'
 'Per 36 Playoffs',
 'Per 100 Poss',
 'Per 100 Poss Playoffs',
 'Advanced',
 'Advanced Playoffs',
 'Adjusted Shooting',
 'Adjusted Shooting Playoffs',
 'Play By Play',
 'Play By Play Playoffs',
 'Shooting',
 'Shooting Playoffs',
 'Game Highs',
 'Game Highs Playoffs',
 'Playoffs Series',
 'All-Star Games'

 ]
```

```python

chrome_options = Options()
chrome_options.add_argument("--headless")
driver = webdriver.Chrome(options=chrome_options)
```

```python
for link in ['https://www.basketball-reference.com/players/j/jordami01.html','https://www.basketball-reference.com/players/j/jamesle01.html']:

    driver.get(link)
    data = pd.read_html(driver.page_source)
    driver.close()

    # print(len(data))
    for index,table in enumerate(table_list):
        if 'jord' in link:

            data[index].to_csv(f'Stats/Jordan/{table}.csv')
        else:
            data[index].to_csv(f'Stats/James/{table}.csv')

```
