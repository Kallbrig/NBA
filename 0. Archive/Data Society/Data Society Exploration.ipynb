#### All Stats, 1 Year

```python

for i in df.select_dtypes(include=[np.number]).columns:

    df.loc[df['Year']==2018,i].plot(kind='hist')
    plt.title(i)
    plt.show()
    break
```

#### By Year

```python
stat = 'PTS'

for year in range(1980,2018):

    df.loc[df['Year']==year,stat].hist()
    plt.title(f'{i} - {stat}')
    plt.show()
```
