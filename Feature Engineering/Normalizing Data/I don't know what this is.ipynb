```python
teams = pd.read_csv('../Data/Team Data/Combined/league_stats.csv',index_col='Unnamed: 0')
```

```python
def fix_years(x):
    return str(x).split('-')[0]
    
```

```python
teams.SEASON = teams.SEASON.apply(fix_years)
```
