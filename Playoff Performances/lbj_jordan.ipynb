```python
# Step 1: Calculate the mean and standard deviation of PTS for each season
seasonal_stats = df.groupby('Year')['PTS'].agg(['mean', 'std']).reset_index()

# Step 2: Filter out the 'Year' and 'PTS' columns for LeBron James and Michael Jordan
lbj_pts = lbj[['Year', 'PTS']]
jordan_pts = jordan[['Year', 'PTS']]


# Step 3: Calculate the Z-Scores for each season for both LeBron and Jordan
def calculate_z_score(player_df, seasonal_stats):
    merged_df = pd.merge(player_df, seasonal_stats, on='Year', how='left')
    merged_df['Z-Score'] = (merged_df['PTS'] - merged_df['mean']) / merged_df['std']
    return merged_df[['Year', 'PTS', 'Z-Score']]


lbj_z_scores = calculate_z_score(lbj_pts, seasonal_stats)
jordan_z_scores = calculate_z_score(jordan_pts, seasonal_stats)

# Step 4: Add Z-Scores as new columns to the original DataFrames
lbj = pd.merge(lbj, lbj_z_scores[['Year', 'Z-Score']], on='Year', how='left')
jordan = pd.merge(jordan, jordan_z_scores[['Year', 'Z-Score']], on='Year', how='left')

# Show the first few rows to confirm
# lbj[['Year', 'PTS', 'Z-Score']].head(), jordan[['Year', 'PTS', 'Z-Score']].head()
```

```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt


# Function to convert the Z-Scores to 2023 equivalent stats
def convert_to_2023_stats(peak_6yr_df, all_stats_df, major_stats):
    for stat in major_stats:
        # Get 2023 season stats
        season_2023_stats = all_stats_df[all_stats_df['Year'] == 2023][stat].agg(['mean', 'std'])
        mean_2023 = season_2023_stats['mean']
        std_2023 = season_2023_stats['std']

        # Convert Z-Scores to 2023 equivalent stats
        peak_6yr_df[f'2023_EQ_{stat}'] = convert_to_2023(peak_6yr_df[f'Z-Score_{stat}'], mean_2023, std_2023)

    return peak_6yr_df

```

```python
def plot_peak_years_comparison(players, all_stats_df, stats_to_compare, ranking_metric='VORP', show_annotations=(True, True), peak_years=6):
    player_dfs_2023 = []

    show_age, show_year = show_annotations

    for player_name in players:
        player_df = all_stats_df[all_stats_df['Player'] == player_name].sort_values(by='Year').reset_index(drop=True)
        peak_nyr = find_peak_years(player_df, all_stats_df, ranking_metric, stats_to_compare, peak_years)

        peak_nyr_2023 = convert_to_2023_stats(peak_nyr, all_stats_df, stats_to_compare)
        player_dfs_2023.append(peak_nyr_2023)

    combined_peak_years_2023 = pd.concat(player_dfs_2023, ignore_index=True)

    for stat in stats_to_compare:
        plt.figure(figsize=(12, 8))
        sns_plot = sns.lineplot(x='Peak Year', y=f'2023_EQ_{stat}', hue='Player', marker='o', data=combined_peak_years_2023)

        for line, player in zip(sns_plot.lines, players):
            player_df = combined_peak_years_2023[combined_peak_years_2023['Player'] == player]
            for x_value, y_value, year, age in zip(line.get_xdata(), line.get_ydata(), player_df['Year'], player_df['Age']):
                annotation_text = f"{y_value:.2f}"
                if show_year:
                    annotation_text += f"\nYear: {int(year)}"
                if show_age:
                    annotation_text += f"\nAge: {int(age)}"
                sns_plot.annotate(annotation_text, (x_value, y_value), textcoords="offset points", xytext=(0,10), ha='center')

        plt.title(f'2023-Equivalent {stat} for Peak {peak_years} Years')
        plt.xlabel('Peak Year')
        plt.ylabel(f'2023-Equivalent {stat}')
        plt.legend(title='Player')
        plt.grid(True)
        plt.show()
```

```python
# Function to find the peak N years based on a given metric and create a new DataFrame for that period
def find_peak_years(player_df, all_stats_df, metric, major_stats, peak_years=6):
    rolling_avg = player_df[metric].rolling(window=peak_years).mean()
    peak_idx = rolling_avg.idxmax()
    peak_nyr = player_df.iloc[peak_idx-(peak_years-1):peak_idx+1].copy()
    peak_nyr['Peak Year'] = ['Year ' + str(i+1) for i in range(peak_years)]

    for year in peak_nyr['Year']:
        for stat in major_stats:  # Now it's defined
            year_stats = all_stats_df[all_stats_df['Year'] == year][stat]
            mean_year = year_stats.mean()
            std_year = year_stats.std()
            peak_nyr.loc[peak_nyr['Year'] == year, f'Z-Score_{stat}'] = (peak_nyr[peak_nyr['Year'] == year][stat] - mean_year) / std_year

    return peak_nyr
```

```python
# Define the list of players you want to compare
players_to_compare = ['Michael Jordan','LeBron James']

# Define the list of stats you want to look at
stats_list = ['PTS','AST','WS','PER','VORP','BPM',]

for i in range(len(stats_list)):
    stats_list[i] = f'{stats_list[i]}_playoffs'


# Define the metric you want to use to find the peak years (default is 'VORP')
ranking_metric = 'PER_playoffs'  # Win Shares

# Define whether you want to show annotations for Age and Year (default is True for both)
annotations = (True, True)

# Define the number of peak years you want to consider (default is 6)
num_peak_years = 10

# Call the function
plot_peak_years_comparison(players=players_to_compare, all_stats_df=all_player_stats, stats_to_compare=stats_list, ranking_metric=ranking_metric, show_annotations=annotations, peak_years=num_peak_years)

```
