```python
import pandas as pd
```

```python

nba = pd.read_csv('player_and_team_stats.csv',index_col='Unnamed: 0')
nba.columns
```

```python
pace = nba.Tm_PACE.to_list()

for i in ['FG','FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT',
       'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF',
       'PTS']:
```

```python
nba[['FG','FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT',
       'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF',
       'PTS']].applymap()
```

```python
for i in ['FG','FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT',
       'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF',
       'PTS']:
    new = list()
    col = nba[i].to_list()
    for j in range(len(col)):
        new.append((col[j]*100)/pace[j])
    nba['PACE_NORM_' + i] = new

    
```
