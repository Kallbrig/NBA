```python
import pandas as pd
import numpy as np
```

```python
# Loading the Player data
players = pd.read_csv('../../Data/Combined Player and Team Data/improved_combined_player_and_team_data.csv',index_col='Unnamed: 0')
```

```python
# This combines all DPOY lists into a single list.
# This will be joined with the main player dataframe on the Player and Age columns

df = pd.read_excel(f'../../Downloads/STATISTICS/STATS/Defensive Player of the Year-1983.xlsx',index_col='Unnamed: 0')
df = df.set_axis(df.iloc[0], axis='columns')
df = df.reset_index().drop(columns='index').drop(index=[0,1]).reset_index().drop(columns='index')
df = df.drop(df[(df['Rank'].isna()) | (df['Rank'] == 'Rank')].index)
df.Rank = [int(str(i).strip('T')) for i in df['Rank'].to_list()]
df['DPOY'] = df['Rank'] == 1
df = df.drop(columns=['Tm','G','MP','PTS','TRB','AST','STL','BLK','FG%','3P%','FT%','WS','WS/48'])


for i in range(1984,2022):
    new_df =pd.read_excel(f'../../Downloads/STATISTICS/STATS/Defensive Player of the Year-{i}.xlsx',index_col='Unnamed: 0')
    new_df = new_df.set_axis(new_df.iloc[0], axis='columns')
    new_df = new_df.reset_index().drop(columns='index').drop(index=[0,1]).reset_index().drop(columns='index')
    new_df = new_df.drop(new_df[(new_df['Rank'].isna()) | (new_df['Rank'] == 'Rank')].index)
    new_df.Rank = [int(str(i).strip('T')) for i in new_df['Rank'].to_list()]
    new_df['DPOY'] = new_df['Rank'] == 1
    new_df = new_df.drop(columns=['Tm','G','MP','PTS','TRB','AST','STL','BLK','FG%','3P%','FT%','WS','WS/48'])
    df = pd.concat([df,new_df])
    
df = df.reset_index().drop(columns='index')

df.columns = ['DPOY_Rank','Player', 'Age', 'DPOY_First', 'DPOY_Pts Won', 'DPOY_Pts Max', 'DPOY_Share', 'DPOY']

df.head(5)
```

```python
# Merges the DPOY list into the players Dataframe
players = players.merge(df,how='left',on=['Player','Age'])
```

```python
# Prints the topmost DPOY share seasons of 1983 - Current
players[['DPOY_Rank','Player','Pos','Year','Age','Tm','PTS','DPOY_Share']].sort_values('DPOY_Share',ascending=False).head(5)
```
