```python
import os

import pandas as pd
import seaborn as sns
```

```python

# Calculating the Diffs for all gamelog pulls
for i in range(1947,2024):
    print(i)

    for j in os.listdir(f'{i}'):

        df = pd.read_csv(f'{i}/{j}', index_col=0)
        df['Year'] = i
        df = df.rename(columns={'Tm_2_ORtg': 'ORtg', 'Tm_2_DRtg': 'DRtg',
 'Tm_2_Pace': 'Pace', 'Tm_2_FTr': 'FTr', 'Tm_2_3PAr': '3PAr', 'Tm_2_TS%': 'TS%', 'Tm_2_TRB%': 'TRB%','Tm_2_AST%':'AST%', 'Tm_2_STL%':'STL%', 'Tm_2_BLK%':'BLK%', 'Tm_1_eFG%':'Off_eFG%', 'Tm_1_TOV%':'Off_TOV%',
'Tm_1_ORB%':'Off_ORB%', 'Tm_1_FT/FGA':'Off_FT/FGA', 'Tm_2_eFG%':'Def_eFG%', 'Tm_2_TOV%':'Def_TOV%', 'Tm_2_DRB%':'Def_DRB%',
'Tm_2_FT/FGA':'Def_FT/FGA'})

        col_list =[k[5:] for k in df.columns if ('Tm_1' in k) and ('Name' not in k)]

        for k in col_list:
            df[f'{k}_Diff'] = df[f'Tm_1_{k}'] - df[f'Tm_2_{k}']

        df.to_csv(f'{i}/{j}')
        df.to_csv(f'all/{j}')
```

```python
df_list = []
for i in os.listdir('all'):
    # print(i)
    df_list.append(pd.read_csv(f'all/{i}',index_col=0))
df = pd.concat(df_list)
df.to_csv('all.csv')
```
