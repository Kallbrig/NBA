```python
for year in range(1947, datetime.now().year + 1):

    for team in master_list[year]:


        if not os.path.exists(f'{base_path}{year}'):
            os.mkdir(f'{base_path}{year}')
        if os.path.exists(f'{base_path}{year}/{year}_{team}.csv'):
            print(year,team,'Already Exists')
            continue



        print(year, team)

    # STARTS HERE

        basic_data = getdata(f'https://www.basketball-reference.com/teams/{team}/{year}/gamelog/')
        adv_data = getdata(f'https://www.basketball-reference.com/teams/{team}/{year}/gamelog-advanced/')

        while 'The owner of this website (www.basketball-reference.com) has banned you temporarily from accessing this website' in basic_data or 'The owner of this website (www.basketball-reference.com) has banned you temporarily from accessing this website' in adv_data:
            print("You've been temporarily banned from accessing basketball reference.\nSleeping 1 hour then continuing.")
            time.sleep(3600)
            basic_data = getdata(f'https://www.basketball-reference.com/teams/{team}/{year}/gamelog/')
            adv_data = getdata(f'https://www.basketball-reference.com/teams/{team}/{year}/gamelog-advanced/')

        df_basic = remove_multiindex(pd.read_html(basic_data)[0])
        df_adv = remove_multiindex(pd.read_html(adv_data)[0])

        merged = pd.merge(left=df_basic, right=df_adv, on=['Rk', 'G', 'Date', 'W/L'])

        merged.columns = merged_columns

        merged['Home/Away'] = merged['Home/Away'].fillna(1).replace('@', 2)
        merged['W/L'] = merged['W/L'].replace('W', 1).replace('L', 2)
        merged['Tm_1_Name'] = team

        merged = merged[~merged.G.isna()]
        merged = merged[~(merged.G == 'G')]

        merged.reset_index(drop=True)

        merged = merged[new_col_names]

        merged.to_csv(f'{year}/{year}_{team}.csv')
        merged.to_csv(f'all/{year}_{team}.csv')

        df = pd.merge(left=df_basic, right=df_adv, on=['Rk', 'G', 'Date', 'W/L'])

        team_list = df.iloc[:, 4].dropna().unique().tolist()
        team_list.remove('Opp')
        team_list.append(team)
        team_list = list(set(team_list))

        master_list[year] = team_list

        time.sleep(8)
```
