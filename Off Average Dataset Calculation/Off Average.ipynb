```python
import pandas as pd
```

```python
nba = pd.read_csv('../players_with_award_voting.csv',index_col= 'Unnamed: 0')
```

```python
def calc_new_avg(num, avg):
    return round(num - avg,3)
    
```

```python
avg_dict = dict()
for i in list(nba.loc[:,'Year'].unique()):
    avg_dict[i] = round(nba[nba['Year'] == i].PTS.mean(),3)
```

```python
l = list()
for i in nba.index:
    l.append(calc_new_avg(num = nba.iloc[i].loc['PTS'], avg = avg_dict[nba.iloc[i].loc['Year']]))
        
    
```

```python
col_list = ['Age', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P',
       '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
       'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS',
       'Tm_AGE', 'Tm_W', 'Tm_L', 'Tm_PW', 'Tm_PL', 'Tm_MOV', 'Tm_SOS',
       'Tm_SRS', 'Tm_ORtg', 'Tm_DRtg', 'Tm_NRtg', 'Tm_PACE', 'Tm_FTr',
       'Tm_3PAr', 'Tm_TS%', 'Tm_eFG%', 'Tm_TOV%', 'Tm_ORB%', 'Tm_FT/FGA',
       'Tm_Op_eFG%', 'Tm_Op_TOV%', 'Tm_DRB%', 'Tm_Op_FT/FGA', 'Win%',]
```

```python
for j in col_list:
    l = list()
    for i in nba.index:
        l.append(calc_new_avg(num = nba.iloc[i].loc[j], avg = avg_dict[nba.iloc[i].loc['Year']]))
    nba['OA_'+ j] = l
    print(j + ' Done')
    
```
