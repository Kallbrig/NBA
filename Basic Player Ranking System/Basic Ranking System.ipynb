```
import pandas as pd
import os
from operator import itemgetter
from tabulate import tabulate
```

```
player_keys = ['Player','Season', 'Age', 'Tm', 'Lg', 'Pos', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%',
       '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%',
       'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', ]

total_keys = player_keys 
```

```
e = [csv for csv in os.walk(os.getcwd())][1][2]

e = [i for i in e]


player_list = []

for person in e:
    value = pd.read_csv(f'player_data/{person}').describe().iloc[1]
    value['Player'] = person.strip()

    pt = 1.2
    three_pt = 1.3
    rebound = 1.1
    assist = (.40*3) + (.60*2)
    turnover = -1.4
    fta = 1.2
    steal = 1.3
    foul = -.80

    two = (value['2PA'] * value['2P%']) * 2 * pt
    three = (value['3PA'] * value['3P%']) * 3 * three_pt
    ft = value['FTA'] * value['FT%']
    ast = value['AST'] * assist
    reb = value['TRB'] * rebound
    stl = value['STL'] * steal
    pf = value['PF'] * foul
    to = value['TOV'] * turnover

    ppm = (two + three + ft + ast + reb +  stl + pf + to)/value['MP']
    player_list.append({'Player':value.Player,'ppm':round(ppm,3), 'pp36':round(ppm*36,3)})

sorted_players = sorted(player_list, key = itemgetter('ppm'),reverse=True)
for i in range(len(sorted_players)):
    player = sorted_players[i]['Player']
    ppm = sorted_players[i]['ppm']
    pp36 = sorted_players[i]['pp36']
```

```

file_list = [csv for csv in os.walk(os.getcwd())][1][2]
stat_list = []
for player_name in file_list:
    value = pd.read_csv(f'player_data/{player_name}')
    player_name_fixed = player_name[0:-4].split('_')
    player_name_fixed = str(player_name_fixed[0].capitalize() + ' ' +player_name_fixed[1].capitalize())
    value['Player'] = player_name_fixed
    stat_list.append(value)
```

# Note:
**Points are not basketball points, but points as calculated by the weights. This approximates the total value of a player.**

- 2 Point Basket Made = 2PM * 2 * 1.2
- 3 Point Basket Made = 3PM * 3 * 1.3
- Free Throws Made = FTM
- Assist = Assists * (.40*3) + (.60*2)
- Turnover = Tov * -1.4
- Steal = Stl * 1.3
- Foul = Pf * -0.8

```
pt = 1.2
three_pt = 1.3
rebound = 1.1
assist = (.40*3) + (.60*2)
turnover = -1.4
steal = 1.3
foul = -.80
```

```

player_list = []
index = 1

for player in stat_list:
    for year in player.index:


        two = (player.iloc[year]['2PA'] * player.iloc[year]['2P%']) * 2 * pt        
        three = (player.iloc[year]['3PA'] * player.iloc[year]['3P%']) * 3 * three_pt
        ft = player.iloc[year]['FTA'] * player.iloc[year]['FT%']
        ast = player.iloc[year]['AST'] * assist
        reb = player.iloc[year]['TRB'] * rebound
        stl = player.iloc[year]['STL'] * steal
        pf = player.iloc[year]['PF'] * foul
        to = player.iloc[year]['TOV'] * turnover

        ppm = (two + three + ft + ast + reb +  stl + pf + to)/player.iloc[year]['MP']

        if not pd.isna(ppm):

            player_list.append({'Rank':index,'Player':player.iloc[year].Player,'ppm':round(ppm,3), 'pp36' : round(ppm*36,3),'ppg' : round(ppm*player.iloc[year].MP,3) , 'Age':player.iloc[year].Age,'Season':player.iloc[year].Season})
            index += 1

```

```
# Sort by Points Per Game 


sorted_players = sorted(player_list, key = itemgetter('ppg'),reverse=True)

for i in range(len(sorted_players)):

    sorted_players[i]['Rank'] = i+1


dataset =sorted_players
header = dataset[0].keys()
rows =  [x.values() for x in dataset[0:50]]
print(tabulate(rows, header,tablefmt='grid'))
```
