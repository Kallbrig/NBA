```python
import pandas as pd
from datetime import datetime
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV, KFold, BaseCrossValidator
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectPercentile
from itertools import product
import os
import pandas as pd

from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.feature_selection import SelectPercentile
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.model_selection import RandomizedSearchCV
```

```python

# Helper function for trimming the dataset
def trim_set(frame: pd.DataFrame) -> pd.DataFrame:
    return frame[((frame['TRB'] > 2) | (frame['AST'] > 5)) & (frame['PTS'] > 15)]

def get_nba_season_year(date=None):
    if date is None:
        date = datetime.now()
    return date.year + 1 if date.month >= 10 else date.year
```

```python

current_season_start_year = get_nba_season_year() - 1
mvp_data_dir = 'Basketball Reference Stat Scraper/Awards/MVP'
player_data_dir = 'Basketball Reference Stat Scraper/player_stats/'

# Read and store data in a list
player_data = [pd.read_csv(os.path.join(player_data_dir, f'{year}_player_stats.csv'),index_col=0).assign(Player=lambda df: df.Player.str.strip('*')) for year in range(1980, current_season_start_year + 1)]
mvp_data = [pd.read_csv(os.path.join(mvp_data_dir, f'{year}_MVP.csv'),index_col=0) for year in range(1980, current_season_start_year + 1)]




# Merge, fill NaNs, and concatenate in one go
full_data = pd.concat([pd.merge(p, m, how='left', on='Player').fillna(0) for p, m in zip(player_data, mvp_data)])

# Drop columns ending with '_y'
full_data = full_data.drop(columns=[col for col in full_data.columns if col.endswith('_y')])

# Rename columns ending with '_x' by removing '_x'
full_data = full_data.rename(columns={col: col.rstrip('_x') for col in full_data.columns if col.endswith('_x')})




# # Prepare features and target without trimming
feature_list = ['Age', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA',
                'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'PER', 'TS%', '3PAr', 'FTr',
                'ORB%', 'DRB%', 'TRB%', 'AST%', 'STL%', 'BLK%', 'TOV%', 'USG%', 'OWS', 'DWS', 'WS', 'WS/48', 'OBPM',
                'DBPM', 'BPM', 'VORP']

x = full_data[feature_list]
y = full_data['Share']
```

```python

class DataTrimmer(BaseEstimator, TransformerMixin):
    def __init__(self, trb_threshold=2, ast_threshold=5, pts_threshold=15):
        self.trb_threshold = trb_threshold
        self.ast_threshold = ast_threshold
        self.pts_threshold = pts_threshold

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        # Apply the trimming logic
        mask = ((X['TRB'] > self.trb_threshold) | (X['AST'] > self.ast_threshold)) & (X['PTS'] > self.pts_threshold)
        X_trimmed = X[mask]
        if y is not None:
            y_trimmed = y[mask]
            return X_trimmed, y_trimmed
        return X_trimmed

```

```python

# Define the preprocessor pipeline
preprocessor = Pipeline([
    ('data_trimmer', DataTrimmer()),  # Use default threshold values
    ('feat_selection', SelectPercentile(percentile=60)),
    ('scaler', MinMaxScaler())
])

# Define the full pipeline with RandomForestRegressor
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', RandomForestRegressor(n_jobs=-1))
])

# Define the parameter grid for RandomizedSearchCV
param_grid = {
    'preprocessor__data_trimmer__trb_threshold': range(0,11,1),
    'preprocessor__data_trimmer__ast_threshold': range(0,11,1),
    'preprocessor__data_trimmer__pts_threshold': range(0,20,2),
    'model__n_estimators': range(100, 501, 100),
    'model__max_depth': range(10, 101, 10),
    'model__min_samples_split': range(2, 11, 2),
    'model__min_samples_leaf': range(1, 11, 1),
    'model__max_features': ['sqrt', 'log2', None]
}

# Create RandomizedSearchCV
random_search = RandomizedSearchCV(
    pipeline, 
    param_distributions=param_grid, 
    n_iter=100, 
    cv=5, 
    verbose=3, 
    random_state=42, 
    n_jobs=-1
)


```

```python
# Run the random search
# Make sure x and y are defined and represent your dataset
random_search.fit(x, y)
```
