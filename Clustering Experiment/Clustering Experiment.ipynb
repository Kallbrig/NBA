```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans

from matplotlib import pyplot as plt

from sklearn.model_selection import train_test_split as tts

from tabulate import tabulate

from sklearn.neighbors import KNeighborsClassifier

pd.set_option('max_columns', None)
```

```python
df = pd.read_csv('Data/Combined Player and Team Data/improved_combined_player_and_team_data.csv',index_col='Unnamed: 0')
df = df.dropna()

df_2018 = df[df.Year == 2000]


df_train = df.drop(df_2018.index)



df_train = df_train.drop(columns=['Player','Pos','Tm'])

df_test_plug_in = df_2018.drop(columns=['Player','Pos','Tm'])
```

```python


kmeans = KMeans(n_clusters=5, random_state=0).fit(df_train)


# df[X.pred_pos == 4].Pos
# df
```

```python
a = kmeans.predict(df_test_plug_in)

kmeans.cluster_centers_

# df_2018
```

```python
df_2018['pred'] = a
```

```python
pd.set_option('display.max_rows', None)
df_2018[['Player','Tm','Pos','MP','PTS','eFG%','AST','TRB','STL','BLK','Tm_W','Tm_L']].loc[df_2018['pred'] == 0].to_csv('zero.csv',index = False)
df_2018[['Player','Tm','Pos','MP','PTS','eFG%','AST','TRB','STL','BLK','Tm_W','Tm_L']].loc[df_2018['pred'] == 1].to_csv('one.csv',index = False)
df_2018[['Player','Tm','Pos','MP','PTS','eFG%','AST','TRB','STL','BLK','Tm_W','Tm_L']].loc[df_2018['pred'] == 2].to_csv('two.csv',index = False)
df_2018[['Player','Tm','Pos','MP','PTS','eFG%','AST','TRB','STL','BLK','Tm_W','Tm_L']].loc[df_2018['pred'] == 3].to_csv('three.csv',index = False)
df_2018[['Player','Tm','Pos','MP','PTS','eFG%','AST','TRB','STL','BLK','Tm_W','Tm_L']].loc[df_2018['pred'] == 4].to_csv('four.csv',index = False)

# tabulate(zero, headers=zero.columns,showindex=False)
```
