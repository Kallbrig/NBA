```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tm = pd.read_csv('/Users/chaseallbright/Dropbox/NBA/Project/Data/teams.csv',index_col=0)
```

```python
tm['SEASON'] = [int(i.split('-')[0]) + 1 for i in tm.SEASON.tolist()]
```

```python
yrs = list(tm['SEASON'].unique())
```

```python
pce = {}

for i in yrs:
    pce[i] = tm[tm['SEASON'] == i].PACE.mean()
```

```python
pce_med = {}
pce_max = {}
pce_min = {}
pce_mea = {}

for i in yrs:
    pce_med[i] = round(tm[tm['SEASON'] == i].PACE.median(),2)
    pce_mea[i] = round(tm[tm['SEASON'] == i].PACE.mean(),2)
    pce_max[i] = round(tm[tm['SEASON'] == i].PACE.max(),2)
    pce_min[i] = round(tm[tm['SEASON'] == i].PACE.min(),2)

ts = {}

for i in yrs:
    ts[i] = round(tm[tm['SEASON'] == i]['TS%'].mean(),2)

threepar = {}

for i in yrs:
    threepar[i] = round(tm[tm['SEASON'] == i]['3PAr'].mean(),2)

FTr = {}

for i in yrs:
    FTr[i] = round(tm[tm['SEASON'] == i]['FTr'].mean(),2)
```

```python
plt.figure(facecolor='white',figsize=(8,6))
plt.legend(['Max', 'Avg','Min'])

plt.xlabel('Year')
plt.ylabel('Pace')
plt.title('League Pace over time')

plt.grid(True)

# plt.plot(yrs, pce_med.values(), 'D-g')
plt.plot(yrs, pce_max.values(), '^-g')
plt.plot(yrs, pce_mea.values(), 'X-k')
plt.plot(yrs, pce_min.values(), 'v-r')

plt.savefig('Lg_Pace.png')
plt.show()
```
